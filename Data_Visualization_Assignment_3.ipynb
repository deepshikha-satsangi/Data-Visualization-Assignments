Plot using sns.relplot, line plot, box plot on fmri data set


```python
import seaborn as sns 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = sns.load_dataset('fmri')
```

```python
data.head()
```

```python
data.tail()
```

```python
data['event'].unique()
```

```python
data['region'].unique()
```

```python
data['timepoint'].unique()
```

```python
sns.relplot(x='timepoint', y='signal', kind = 'line', data = data)
```

```python
sns.catplot(x='event', y='signal', data = data)
```

```python
sns.relplot(x='timepoint', y='signal', kind= 'scatter', data = data, legend = 'full')
```

```python
sns.catplot(x='event', y='signal', data = data, kind = 'box')
```

```python
sns.catplot(x='event', y='signal', data = data, kind= 'violin')
```

```python
sns.catplot(x='event', y='signal', data=data, kind= 'boxen')
```

```python
sns.catplot(x='signal', y='event', data = data, kind = 'box', orient = 'horizontal')
```

```python
sns.catplot(x='event', y='signal', data = data, kind = 'swarm')
```
