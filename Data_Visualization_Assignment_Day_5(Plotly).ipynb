QUESTION 3 - plot a Pie Chart of Indian Car companies market share % values (max 6 Companies market share )


```python
import matplotlib.pyplot as plt

cars=['Ford', 'Toyota', 'Honda', 'Maruti Suzuki', 'Tata Motors', 'Renault Triber']
market_share=[24111, 24567, 35330, 373120, 274621, 21963]

plt.pie(market_share,labels=cars,shadow=True)
plt.axis("equal")
plt.legend(title="Share Values of Cars")
plt.show()
```

QUETION 4 -Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )


```python
import matplotlib.pyplot as plt
x1=['UP', 'MP', 'Bihar', 'Assam', 'Kerala', 'Karnataka', 'Jharkhand', 'Delhi', 'Chhattisgarh', 'Chandigarh']
y1=[1707225,790152,723147,532084,3053116,2869320,346279,1435030,997785,61823,]
plt.bar(x1, y1, label="Yellow Bar", color='y')
plt.plot()

plt.xlabel("States")
plt.ylabel("Active Cases")
plt.title("Bar Chart of Covid-19 Active Cases")
plt.legend()
plt.show()
```
